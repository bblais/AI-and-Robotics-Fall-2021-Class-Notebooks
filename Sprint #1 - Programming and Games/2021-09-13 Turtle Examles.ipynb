```python
reset(figsize=(5,5))  # for the class notebooks, including figsize

size=80

print("start")
for i in range(4):
    forward(size)
    right(90)
    print("here")
print("end")

penup()
forward(200)  # move over
pendown()

size=30
print("start")
for i in range(4):
    forward(size)
    right(90)
    print("here")
print("end")
```

```python
def draw_square(size):
    for i in range(4):
        forward(size)
        right(90)
    
def move_over(size):
    penup()
    forward(size)  # move over
    pendown()
```

```python
reset(figsize=(5,5))  # for the class notebooks, including figsize

draw_square(80)

move_over(200)

draw_square(30)

```
